#Q9.Top 5 Customers

#Q11.Orders by Payment Mode

```python
payment_mode = df['PaymentMethod'].value_counts()

payment_mode
```

#12.Monthly Sales Trend

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['Month'] = df['OrderDate'].dt.month_name()
```

```python
df['Month'] = df['OrderDate'].dt.to_period('M')

monthly_sales = df.groupby('Month')['TotalAmount'].sum()

monthly_sales
```

#Q13.Sales by State and Category

```python
state_category_sales = df.groupby(
    ['State','Category']
)['TotalAmount'].sum()

state_category_sales
```

#Q15.Bar Chart – Sales by State

#Q16.Pie Chart – Category-wise Sales

```python
import matplotlib.pyplot as plt

# Category-wise sales
category_sales = (
    df.groupby('Category')['TotalAmount']
      .sum()
      .sort_values(ascending=False)
)

colors = [
    '#264653',
    '#2A9D8F',
    '#E9C46A',
    '#F4A261',
    '#E76F51',
    '#8E44AD'
]

fig, ax = plt.subplots(figsize=(10, 8))

# Aesthetic background color
bg_color = '#DCE8E5'
fig.patch.set_facecolor(bg_color)
ax.set_facecolor(bg_color)

# Donut Chart
wedges, texts, autotexts = ax.pie(
    category_sales,
    labels=category_sales.index,
    colors=colors,
    startangle=90,
    autopct='%1.1f%%',
    pctdistance=0.82,
    labeldistance=1.15,
    wedgeprops={
        'width': 0.38,
        'edgecolor': 'white',
        'linewidth': 2
    },
    shadow=True
)

# Percentage text style
for t in autotexts:
    t.set_color('white')
    t.set_fontsize(14)
    t.set_fontweight('bold')

# Label style
for t in texts:
    t.set_fontsize(15)
    t.set_fontweight('bold')
    t.set_color('#0B1F33')

# Center circle
centre_circle = plt.Circle(
    (0, 0),
    0.62,
    fc='#C9E0DD',
    ec='lightgray',
    linewidth=3
)
ax.add_artist(centre_circle)

# Center text
ax.text(
    0,
    0,
    f"₹{category_sales.sum()/1e6:.1f}M\nSales",
    ha='center',
    va='center',
    fontsize=28,
    fontweight='bold',
    color='#0B1F33'
)

plt.title(
    'Category-wise Sales',
    fontsize=24,
    fontweight='bold',
    color='#0B1F33',
    pad=25
)

plt.tight_layout()
plt.show()
```
